```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# import polars
import polars as pl
```

```python
df = pl.read_csv("vehicle.csv")
```

```python
cols = df.columns
cols.remove('class')
```

```python
for i in np.arange(len(cols)):
    sns.histplot(df[cols[i]], color='blue')
    plt.show()
    print('Distribución de ',cols[i])
    print('Promedios es:', df[cols[i]].mean())
    print('Mediana es:',df[cols[i]].median())
    print('Moda es:',df[cols[i]].mode())
    print('Desviación estándar es:',df[cols[i]].std())
    print('Oblicuidad es:',df[cols[i]].skew())
    print('Máximos es:',df[cols[i]].max())
    print('Mínimo es:',df[cols[i]].min())
```
